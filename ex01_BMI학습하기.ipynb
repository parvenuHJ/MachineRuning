```python
data = pd.read_csv('bmi_500.csv')
data
```

```python
# 문제데이터, 답데이터로 분리
# input_data 라는 변수에는 문제 데이터 담기('Height : Weight')
# target_data 라는 변수에는 답 데이터 담기('Label')

input_data = data.loc[:,'Height':'Weight']
target_data = data.loc[:,'Label']
```

```python
# 문제와 답데이터의 모양/크기 확인하기
# shape 메소드 활용 -> 행과 열값 확인가능
print(input_data.shape) # -> 문제데이터 : 2차원으로
print(target_data.shape) # -> 답데이터 : 1차원으로
```

```python
# iloc -> index 번호로 접근할거냐
# loc -> 이름으로 접근할거냐 
# loc [행 , 열] ~> loc [1:3, 1:4]

# 훈련세트와 테스트세트로 나누기
# 훈련세트 : 테스트세트 = 7 : 3
X_train = input_data.iloc[:350, :]
X_test = input_data.iloc[350: , :]
y_train = target_data.iloc[:350]
y_test = target_data.iloc[350:]
```

```python
# 각각의 데이터의 크기/모양 확인
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 사이킷런 라이브러리 가져오기
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics # 평가를 위한 모듈
# .fit() 으로 훈련할 수 있음
```

```python
# 모델 생성
# 하이퍼파라미터 튜닝
knn_model = KNeighborsClassifier(n_neighbors=10) # 이웃을 10개를보겠다 -> 근접한 데이터 10개를 보고 분류진행
knn_model
```

```python
# 훈련 진행 fit 메소드
knn_model.fit(X_train, y_train)
```

```python
# 예측 predict 메소드
pre = knn_model.predict(X_test)
pre
```

```python
y_test
```

```python
# 정확도 측정 -> metrics (측정할값, 정답)
metrics.accuracy_score(pre,y_test)
```

## K-Nearest Neighbors : k-최근접 이웃 알고리즘

- k가 너무 높으면 -> 훈련데이터에 관계없이 과반에 의해 결정됨 => 과소적합 
- k가 너무 낮으면 -> 훈련데이터가 어떻게 생겼냐에 따라에 너무 과도하게 결정됨 => 과대적합[ 모델의 복잡도가 증가 ]


- 과대 : 훈련데이터에 너무 적합 
- 과소 : 훈련데이터와 크게 영향X
