### 정제(cleaning)
- 코퍼스에는 아무 의미도 없거나 목적에 부합하지 않은 단어들도 포함
- 전처리 과정에서 토큰들을 제거하는 작업을 **정제**
- 등장빈도, 단어길이, 불용어 등을 기준으로 사용

```python
# 등장빈도가 2 이하인 당너들만 추출

# 단어의 빈도 계산에 사용할 Counter() 함수 사용
from collections import Counter
```

```python
# 전체 단어 토큰 리스트 = tokenized_words

# Counter 함수를 통해 단어의 빈도수를 카운트하여 집합을 생성
vocab = Counter(tokenized_words)
```

```python
# 빈도수가 2 이하인 단어 리스트를 추출

# items() 함수 활용 - 키와 값들의 쌍을 얻을 수 있게 해준다.
# (1) 리스트 컴프리핸션
low_freq_words = [key for key, value in vocab.items() if value <=2]
```

```python
# 빈도수가 2 이하인 단어 리스트를 추출
# (2) 반복문을 활용해서 빈도수가 2 이하인 단어 리스트를 추출
# vocab - 딕셔너리 형태 -> for key, value in vocab.items(): 이런 형태로 뽑아서 print(key, value) 이렇게 뽑을 수 있음
# low_freq_words - 리스트명

low_freq_words = []
for key, value in vocab.items():
    # print(key, value)
    if value <3 :
        low_freq_words.append(key)
low_freq_words
```

```python
# 등장 빈도가 2 이하인 단어 리스트 개수확인
print("빈도수가 2이하인 단어수 : ", len(low_freq_words))
```

```python
# 빈도수가 2 이하인 단어들을 제거한 결과를 따로 저장
# 리스트명 : cleaned_by_freq
# not in, 반복문, 조건문 사용
cleaned_by_freq = []

for word in tokenized_words: # tokenized_words (토큰화된 단어들 리스트)에 word가
    if word not in low_freq_words:# low_freq_words(빈도수 2이하인 단어들)에 없으면 /빈도수가 3이상이니까
        cleaned_by_freq.append(word) #cleaned_by_freq 리스트에 추가해라
cleaned_by_freq
```

```python
print("빈도수가 3이상인 단어수 : " , len(cleaned_by_freq))
```

```python
# 길이가 2 이하인 단어 제거
# 콤마, i, be, 등 의미를 갖지 않는 단어 제거
cleaned_by_freq_len = []

# cleaned_by_freq 에서 길이가 2 이하인 단어를 제거
# 제거된 값은 cleaned_by_freq_len 이라는 변수에 담기
for word in cleaned_by_freq:
    if len(word) >= 3 :
        cleaned_by_freq_len.append(word)
```

```python
# 등장 빈도 기준 정제 함수 정의
def clean_by_freq(tokenized_words, freq):
    # 1. Counter 함수를 통해 단어의 빈도수를 카운트하여 단어 집합 생성
    # vocab 이라는 변수에 담기
    vocab = Counter(tokenized_words)

    # 2. 빈도수가 freq 이하인 단어 추출
    # low_freq_words 라는 변수에 담기
    low_freq_words = []
    for key, value in vocab.items():
        if value <= freq:
            low_freq_words.append(key)

    # 3. low_freq_words 에 포함되지 않는 단어 리스트 생성
    # cleaned_words 라는 변수에 담기
    cleaned_words = []
    for word in tokenized_words:
        if word not in low_freq_words:
            cleaned_words.append(word)

    return cleaned_words

# 단어 길이 기준 정제 함수
def clean_by_len(tokenized_words, length):
    cleaned_by_freq_len = []
    # 단어 길이가 length 이상인 단어들을 cleaned_by_freq_len에 담아주기
    for word in tokenized_words:
        if len(word) >= length:
            cleaned_by_freq_len.append(word)

    return cleaned_by_freq_len
```

### 불용어 처리 함수 만들기

```python
# 불용어 제거 함수 만들기
def clean_by_stopwords(tokenized_words, stop_words_set):
    
    cleaned_words = []
    
    # 불용어를 제거해주는 코드 작성
    for word in tokenized_words:
        if word.lower() not in stopwords_set :
            cleaned_words.append(word)
    
    return cleaned_words
```

### 정규화 방법 2. 규칙 기반 정규화

```python
text = "She became a US citizen. Ummmm, I think, maybe and or."

# 동의어 사전을 만들어서 정규화진행
# US를 USA, Ummm을 Umm 으로 통합 
synonym_dict = {'US' : 'USA', 'U.S' : 'USA', 'Ummmm' : 'Umm', 'Ummm': 'Umm'} # key - 바꿔야할것 # value - 통합

# 단어 토큰화 (word_tokenize 함수 사용)
# tokenized_words 에 담아주기
tokenized_words = word_tokenize(text)

# 동의어 사전에 있는 단어라면, value에 해당하는 값으로 반환
normalized_words = []
for word in tokenized_words:
    if word in synonym_dict.keys(): # synonym_dict = 딕셔너리 형태 / 이 딕셔너리의 keys - .keys()
        word = synonym_dict[word] # 딕셔너리에서 synonym_dict['US'] 이런식으로 key값을 적으면 value 값을 얻어낼 수 있음!!!!!!!!!!!!
    
    normalized_words.append(word)

print(normalized_words)
```

### 어간추출 함수 코드화

```python
from nltk.stem import PorterStemmer

# 포터스테머 어간 추출함수
def stemming_by_porter(tokenized_words):
    porter_stemmer = PorterStemmer()
    porter_stemmer_words = []
    
    for word in tokenized_words:
        stem = porter_stemmer.stem(word)
        porter_stemmer_words.append(stem)
    
    return porter_stemmer_words
```

### 품사 태깅(POS : Part of Speech Tagging)
- 각 단어가 어떤 품사로 쓰였는지 표시하는 작업 = 품사태깅
- 여러문장코퍼스 -> 문장 토큰화 -> 단어 토큰화 -> 품사 태깅

```python
# 품사 태깅을 할 코퍼스를 문장 기준으로 토큰화
text = "Watching Time Chasers, it obvious that it was made by a bunch of friends. Maybe they were sitting around one day in film school and said, \"Hey, let\'s pool our money together and make a really bad movie!\" Or something like that."

tokenized_sents = sent_tokenize(text)
```

```python
# 토큰화된 문장들을 반복하면서 단어 토큰화와 품사 태킹 작업 진행
pos_tagged_list = []

for sentence in tokenized_sents:
    # 단어 토큰화
    tokenized_words = word_tokenize(sentence)
    # 품사 태깅
    pos_tagged = pos_tag(tokenized_words)
    # pos_tagged_list에 값 담아주기
    pos_tagged_list.append(pos_tagged)
```

```python
# [(단어, 품사태그), (단어, 품사태그)] 형태로 품사 태깅된 결과 반환
# Penn Treebank POS Tags
print(pos_tagged_list)
```

### 품사 태깅 함수 만들기

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tag import pos_tag

# 품사 태깅 함수 정의
def pos_tagger(tokenized_sents):
    pos_tagged_words = []
    # 단어 토큰화 
    for sentence in tokenized_sents:
        tokenized_words = word_tokenize(sentence)
    # 품사 태깅
        pos_tagged = pos_tag(tokenized_words)
    # 품사 태깅한 데이터 담아주기(extend)
        pos_tagged_words.extend(pos_tagged)
    return pos_tagged_words
```

### WordNet Pos Tag : 영어 어휘 데이터베이스에 적용되어 있는 품사태그
- N으로 시작하면 wn.NOUN
- J로 시작하면 wn.ADJ
- R로 시작하면 wn.ADV
- V로 시작하면 wm.VERB

```python
from nltk.corpus import wordnet as wn
```

```python
# pennTreePosTag를 WordNetPosTag로 변환
def penn_to_wn(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    else:
        return
```

```python
# 표제어 추출을 위한 함수 호출
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# 표제어 추출하기 위한 WordNetLemmatizer 클래스에 있는 lemmatizer() 함수 호출
lemmatizer = WordNetLemmatizer()
lemmatizer_words = [] # 표제어가 추출된 단어를 담을 리스트

for word, tag in tagged_words:
    # WordNet Pos Tag 로 변환하는 함수 호출
    wn_tag = penn_to_wn(tag)
    
    # 품사를 기준으로 표제어 추출
    if wn_tag in (wn.NOUN, wn.ADJ, wn.ADV, wn.VERB):
        lemmatizer_words.append(lemmatizer.lemmatize(word, wn_tag)) # 표제어 추출 함수
    else :
        lemmatizer_words.append(word)
```

### 표제어 추출함수 정의

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet as wn
nltk.download('wordnet')
nltk.download('omw-1.4')
# pennTree -> WordNet으로 변환
def penn_to_wn(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    else:
        return

def words_lemmatizer(pos_tagged_words): # 표제어를 추출해주는 함수
    lemmatizer = WordNetLemmatizer() # 객체 생성
    lemmatized_words = [] # 표제어 추출된 단어를 담는 리스트
    
    for word, tag in pos_tagged_words:
        wn_tag = penn_to_wn(tag)
        
        if wn_tag in (wn.NOUN, wn.ADJ, wn.ADV, wn.VERB):
            lemmatized_words.append(lemmatizer.lemmatize(word, wn_tag))
        else :
            lemmatized_words.append(word)
        return lemmatized_words
```
