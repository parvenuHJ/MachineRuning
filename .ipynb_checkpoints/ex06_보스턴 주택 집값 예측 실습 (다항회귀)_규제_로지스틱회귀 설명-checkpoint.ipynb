```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#   
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
#         
# from sklearn.datasets import load_boston
```

```python
#    
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows = 22, header = None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2] # 1::2 -> 1  2 
```

```python
data
```

```python
# data numpy   -> pd.DataFrame  
boston = pd.DataFrame(data, columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
boston
```

```python
boston.shape
```

```python
#      -> train_test_split()
X_train, X_test, y_train, y_test = train_test_split(boston, target, test_size = 0.3)
```

```python
#   
plt.figure(figsize = (5,5))
plt.scatter(X_train['LSTAT'], y_train, label = 'original')

#   
LSTAT_model = LinearRegression() #  
LSTAT_model.fit(X_train[['LSTAT']], y_train) #   -> LSTAT   -> !!!)  2!!

#   
pre = LSTAT_model.predict(X_train[['LSTAT']]) #  2
plt.scatter(X_train['LSTAT'], pre, label = '', color = 'red')

plt.legend()
plt.xlabel('LSTAT') # 
plt.ylabel('PRICE')

plt.show()
```

```python
#   
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows = 22, header = None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2] # 1::2 -> 1  2 
```

```python
boston = pd.DataFrame(data, columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(boston, target, test_size = 0.3)
```

```python
lr_model = LinearRegression()
```

```python
# 
lr_model.fit(X_train, y_train)
```

```python
# 
pre = lr_model.predict(X_test)
pre
```

```python
# (mse, rmse)
# mse (mean squared error) -   
mse = mean_squared_error(pre, y_test)
print("  mse : ", mse)

# rmse -   
rsme = mse**0.5
print("  rmse :", rsme)
```

#### L2(Ridge)

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge() #  =  ( = 1.0 )
```

```python
ridge_model.fit(X_train, y_train)
```

```python
# 
ridge_pre = ridge_model.predict(X_test)
ridge_pre
```

```python
# 
# mse (  )
mse = mean_squared_error(ridge_pre, y_test)
# rmse (  )
rmse = mse ** 0.5

# r2

# r2   
from sklearn.metrics import r2_score
r2 = r2_score(ridge_pre, y_test)

print('ridge  mse  :', mse)
print('ridge  rmse  :', rmse)
print('ridge  r2  :', r2)
```

#### L1(Lasso)

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso() # ()  - 1.0
```

```python
lasso_model.fit(X_train, y_train)
```

```python
lasso_pre = lasso_model.predict(X_test)
```

```python
#  (mse, rmse, r2)
mse = mean_squared_error(lasso_pre, y_test)

rsme = mse ** 0.5

r2 = r2_score(lasso_pre, y_test)

print("Lasso mse :" , mse)
print("Lasso rmse :" , rmse)
print("Lasso r2 :" , r2)
```

### Ridge vs Lasso

```python
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
ridge_coef_list = []
lasso_coef_list = []

for a in alpha_list :
    r_model = Ridge(alpha = a)
    l_model = Lasso(alpha = a)
    r_model.fit(X_train, y_train)
    l_model.fit(X_train, y_train)
    ridge_coef_list.append(r_model.coef_)
    lasso_coef_list.append(l_model.coef_)
```

```python
ridge_coef_list # ridge       
```

```python
lasso_coef_list # lasso       
```

```python
#    
# ridge    0  0  
rid_df = pd.DataFrame(np.array(ridge_coef_list).T, columns = alpha_list) # .T ->   
rid_df
```

```python
#    
# lasso    0   
# L1 ()      .
lasso_df = pd.DataFrame(np.array(lasso_coef_list).T, columns = alpha_list) # .T ->   
lasso_df
```
