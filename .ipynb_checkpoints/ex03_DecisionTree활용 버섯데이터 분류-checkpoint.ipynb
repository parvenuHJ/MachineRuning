###  :   (ex) u-1, g-2 ,...)
->  (key : value )  

```python
#  
habitat_dict = {
    'u' : 2,
    'g' : 1,
    'm' : 3,
    'd' : 5,
    'p' : 4,
    'w' : 6,
    'l' : 7
}
habitat_dict
```

```python
# mapping -> .map()  
X['habitat'].map(habitat_dict)
```

#   2 : LableEncoder (sklearn)

```python
from sklearn.preprocessing import LabelEncoder

items = X['habitat'].unique()

# LabelEncoder  
encoder = LabelEncoder()

#.fit() :   
encoder.fit(items)

# .transfrom() :     
labels = encoder.transform(items)

print('  :' ,labels)
```

```python
print(' :', encoder.classes_)
```

```python
# get_dummies()   -> pandas  
X_one_hot = pd.get_dummies(X) # 0 1   
X_one_hot
```

```python
#      
#  -> 
print("    : \n", X.columns)
print("    : \n", X_one_hot.columns)
```

# 

```python
# train_test_split 
from sklearn.model_selection import train_test_split
# (X_one_hot)  (y)    ( 7 : 3 )  
X_train, X_test, y_train, y_test = train_test_split(X_one_hot,
                                                            y,
                                                            test_size = 0.3)
```


- max_depth = 3 =>      

```python
#  <-  
from sklearn.tree import DecisionTreeClassifier
#    - tree   
tree = DecisionTreeClassifier(max_depth = 3)
tree
```

# `  ` = ` `
## ->    ()  


```python
#  - .fit()  
tree.fit(X_train, y_train)
```

```python
#  - predict()  
pre = tree.predict(X_test)
```

```python
#  

#   
from sklearn import metrics

# metrics / accuracy_score  
print("  : " , metrics.accuracy_score(pre, y_test))
```
